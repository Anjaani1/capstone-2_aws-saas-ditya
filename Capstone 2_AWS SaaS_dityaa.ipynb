### • Penambahan Kolom Baru(Profit Margin)
Untuk mendukung analisis, akan ditambahkan kolom Profit Margin yang menunjukkan persentase keuntungan dari total penjualan (Sales). Kolom ini membantu melihat seberapa efisien perusahaan mengubah penjualan menjadi laba. Untuk menghitung total profit dan rata-rata profit margin untuk setiap produk. Profit margin dihitung dengan rumus:

\\[
\text{Profit Margin} = \frac{\text{Profit}}{\text{Sales}} \times 100
\\]

```python
df["Profit Margin"] = round((df["Profit"] / df["Sales"]) * 100, 2)
# Display
df.head()
```

#### • Quantity

```python
min_qty = df['Quantity'].min()
max_qty = df['Quantity'].max()

quantity_info = pd.DataFrame({'Least Quantity' : [f'{min_qty} pcs'], 'Most  Quantity' : [f'{max_qty} pcs']})
quantity_info
```

Untuk Quantity menujukkan bahwa pada `Jumlah paling sedikit` maupun `Jumlah paling banyak` yang dibeli oleh Customer tidak ada nilai negatif, artinya bahwa tidak ada anomali pada kolom tersebut

```python
#lihat transaksi dengan Sales terendah dan tertinggi dengan Quantity nya
low_sales = df[df['Sales'] == df['Sales'].min()][['Sales','Product', 'Quantity']]
high_sales = df[df['Sales'] == df['Sales'].max()][['Sales','Product', 'Quantity']]

low_sales['Sales'] = low_sales['Sales'].apply(lambda x: f"${x:,.2f}")
low_sales['Quantity'] = low_sales['Quantity'].apply(lambda x: f"{x} pcs")

high_sales['Sales'] = high_sales['Sales'].apply(lambda x: f"${x:,.2f}")
high_sales['Quantity'] = high_sales['Quantity'].apply(lambda x: f"{x} pcs")

# Tambahkan kolom label
low_sales['Label'] = 'Lowest Sales'
high_sales['Label'] = 'Highest Sales'

# Gabungkan dan set kolom Label jadi index
sales_qty = pd.concat([low_sales, high_sales])
sales_qty = sales_qty[['Label','Product', 'Sales', 'Quantity']].set_index('Label')
sales_qty
```

Dari Tabel di atas menunjukkan **Transaksi dengan penjualan terendah** terjadi pada produk `OneView`, dengan nilai sebesar `$0.44` dan kuantitas `1 pcs`.  **Transaksi dengan penjualan tertinggi** berasal dari produk `Big Ol Database`, dengan nilai mencapai `$22,638.48` hanya untuk `6 pcs`.

### • Pengecekan Dataset untuk Outlier
Sebelum melanjutkan ke tahap analisis lebih lanjut, perlu dilakukan pengecekan terhadap adanya outlier pada data numerik. Outlier merupakan nilai yang secara signifikan menyimpang dari distribusi umum dan dapat memengaruhi hasil analisis karena mewakili nilai yang jauh berbeda dari mayoritas data. Oleh karena itu, sebagai langkah awal dalam memastikan kualitas data, dilakukan visualisasi menggunakan boxplot pada kolom-kolom numerik seperti `Sales`, `Quantity`, `Discount`, dan `Profit` untuk mengidentifikasi keberadaan nilai-nilai ekstrem atau anomali yang mungkin ada dalam dataset.

```python
columns_numerical = ['Sales', 'Quantity', 'Discount', 'Profit']

plt.figure(figsize=(10,8))
for i, col in enumerate(columns_numerical, 1):
    plt.subplot(5, 2, i)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

Boxplot di atas menjelaskan:
1. Boxplot Sales
    - Data Sales punya banyak nilai yang menyebar lebar, dari nilai kecil sampai sangat besar (lebih dari 20.000).
    - Ada banyak outlier di sisi kanan (nilai sangat tinggi di atas kuartil atas), yang artinya ada beberapa transaksi dengan penjualan yang jauh lebih besar dari transaksi rata-rata.
    - Sebagian besar data terkonsentrasi di bagian bawah (nilai kecil sampai menengah).

2. Boxplot Quantity
    - Quantity (jumlah barang yang terjual) berkisar mayoritas antara 1 sampai 9.
    - Ada beberapa outlier di sebelah kanan, yaitu transaksi yang menjual barang dalam jumlah lebih dari 9, sampai 14.
    - Median dan kuartil terletak di sekitar 2-5 barang.

3. Boxplot Discount
    - Mayoritas transaksi memiliki diskon rendah, dengan kuartil atas sekitar 0.2 (20%).
    - Ada beberapa outlier diskon tinggi yang mencapai sampai 0.8 (80%).
    - Diskon besar ini termasuk langka (outlier), dan sebagian besar transaksi diskonnya rendah.

4. Boxplot Profit
    - Profit memiliki distribusi yang cukup simetris tapi dengan beberapa nilai ekstrim (outlier) di sisi negatif dan positif.
    - Ada transaksi yang menyebabkan kerugian besar (profit negatif ekstrim sampai -6000 lebih), serta beberapa transaksi untung sangat besar sampai di atas 8000.
    - Sebagian besar profit berada di kisaran yang lebih dekat ke nol, menunjukkan bahwa banyak transaksi dengan margin keuntungan kecil atau rugi ringan

```python
#define the columns
columns_numerical = ['Sales', 'Quantity', 'Discount', 'Profit']

# initialize an empty list to store the outliers result
outliers_results = []

for column in columns_numerical:
    Q1 = df[column].quantile(0.25)  # First quartile
    Q3 = df[column].quantile(0.75)  # Third quartile
    IQR = Q3 - Q1  # Interquartile range

    # Calculate bounds
    lower_whisker = round(Q1 - 1.5 * IQR)
    upper_whisker = round(Q3 + 1.5 * IQR)

    # Count data points outside bounds
    outliers_below = df[column][df[column] < lower_whisker].count()
    outliers_above = df[column][df[column] > upper_whisker].count()

    # Append the results
    outliers_results.append({
        'Column': column,
        'Lower Whisker': lower_whisker,
        'Upper Whisker': upper_whisker,
        'Outliers Below': outliers_below,
        'Outliers Above': outliers_above,
        'Total Outliers': outliers_below + outliers_above
    })

# Convert the results to a DataFrame
results_df = pd.DataFrame(outliers_results)

# display the resulting table
results_df
```

Tabel di atas mengidentifikasi sebaran outlier pada setiap variabel numerik. Selanjutnya, dilakukan analisis untuk menilai karakteristik dan pengaruh outlier tersebut.

### - Uji Hipotesis antara Segmentasi dan Profit
Dikarenakan yang dibandingkan variabel bersifat kategorik bukan numerik kontinu yaitu Variabel Segment(segmentasi pelanggan) dengan lebih dari dua grup kategori maka menggunakan Kruskal-Wallis test (bukan ANOVA ataupun t-test). Kruskall wallis adalah uji non-parametrik, artinya tidak mengasumsikan distribusi normal dan untuk melihat apakah rata-rata profit berbeda secara signifikan di seluruh segmen.

### - Kruskal Wallis-Test 

`Asumsi` : Menguji apakah ada perbedaan Median Profit antara Segmen Pelanggan SMB, Strategic, dan Enterprise untuk penjualan Produk Marketing Suite ?

`Hipotesis`:

$H_{0} : M_{SMB}=M_{Strategic}=M_{Enterprise}$

$H_{a} : M_{SMB}≠M_{Strategic} \quad atau\quad M_{Enterprise}≠M_{ideal} \quad atau\quad M_{Strategic}≠M_{Enterprise} $

`Signifikansi` -> $\alpha = 0.05$

`Pengambilan Kesimpulan` :
- p-value $\leq \alpha$ : Reject $H_{0}$ $\quad (Ada \quad setidaknya \quad satu \quad pasang \quad Segmen \quad Pelanggan \quad yang \quad memiliki\quad  median\quad  harga\quad  yang\quad  berbeda)$
- p-value $ > \alpha$ : Failed to Reject $H_{0}$ $\quad (Median\quad Profit\quad ketiga\quad Segmen\quad Pelanggan\quad tersebut\quad sama)$

```python
from scipy.stats import kruskal

# Definisikan segmen yang akan diuji
segments = ['SMB', 'Strategic', 'Enterprise']

# Filter data hanya untuk segmen yang diinginkan
filtered_df = df_msuite[df_msuite['Segment'].isin(segments)& (df_msuite['Profit'] < 0)]# karena ingin lihat yang rugi maka difilter profit<0

# Ambil data profit untuk masing-masing segmen
smb_profit = filtered_df[filtered_df['Segment'] == 'SMB']['Profit']
strategic_profit = filtered_df[filtered_df['Segment'] == 'Strategic']['Profit']
enterprise_profit = filtered_df[filtered_df['Segment'] == 'Enterprise']['Profit']

# Lakukan uji Kruskal-Wallis
h_stat, p_value = kruskal(smb_profit, strategic_profit, enterprise_profit)

# Tampilkan hasil
print(f"Kruskal-Wallis Test:\nH-statistic: {h_stat:.3f} \nP-value: {p_value:.4f}")
```

```python
# Interpretasi hasil
if p_value <= 0.05:
    print(f"P-value = {p_value:.4f}, karena p-value <= 0.05, maka kita menolak H0 (ada perbedaan median profit antara setidaknya dua segmen)")
else:
    print(f"P-value = {p_value:.4f}, karena p-value > 0.05, maka kita gagal menolak H0 (tidak ada perbedaan signifikan pada median profit ketiga segmen)")
```

Berdasarkan hasil uji hipotesis, disimpulkan bahwa tidak terdapat perbedaan yang signifikan dalam median profit antar segmen pelanggan (SMB, Strategic, dan Enterprise).

**Nilai p yang diperoleh sebesar 0.098, lebih besar dari tingkat signifikansi 0.05. Artinya, kita tidak menolak hipotesis nol (H0) yang menyatakan bahwa median profit ketiga segmen tersebut tidak berbeda secara signifikan**.

Hasil uji Kruskal-Wallis ini menunjukkan bahwa pada transaksi yang menghasilkan kerugian (profit negatif), median profit dari segmen SMB, Strategic, dan Enterprise relatif serupa, tanpa ada segmen yang secara statistik menunjukkan kinerja jauh lebih buruk atau lebih baik dibandingkan yang lain.

Dengan kata lain, meskipun ketiga segmen mengalami profit negatif, tingkat median kerugiannya tidak berbeda secara signifikan, menandakan bahwa mereka mungkin menghadapi kondisi profitabilitas yang kurang lebih sebanding

### - Analisa Korelasi Antara Sales, Discount, Profit Margin, dan Profit
Analisis ini mengevaluasi hubungan antara Sales, Discount, Profit Margin, dan Profit pada produk Marketing Suite untuk memahami pengaruh diskon terhadap penjualan dan profitabilitas secara keseluruhan.

```python
numerical_col = ["Sales", "Discount", "Profit", "Profit Margin",'Quantity']
correlation = round(df_msuite[numerical_col].corr(method = "spearman"),5)
correlation
```

```python
plt.figure(figsize = (15, 10))
sns.heatmap(
    data = correlation,
    annot = True,
    fmt = ".2f",
    cmap = "YlGnBu",
    center = 0,
)

plt.title("Korelasi Antara Sales, Discount, Profit Margin, dan Profit", size=16, pad=10, fontweight='bold')
plt.show()
```

Gambar heatmap di atas menunjukkan hubungan korelasi antar variabel numerik yang dianalisis terhadap Profit (keuntungan), menggunakan metode korelasi Spearman. Warna pada heatmap menggambarkan seberapa kuat dan arah hubungan antar variabel:
- Warna gelap ke biru = korelasi positif kuat
- Warna terang kekuningan = korelasi negatif kuat
- Warna hijau muda hingga aqua = korelasi lemah atau hampir tidak ada

```python
for col in numerical_col:
    corr = df_msuite[col].corr(df_msuite["Profit"], method="spearman")

    # Tentukan arah korelasi
    if corr > 0:
        direction = "positif"
    else:
        direction = "negatif"

    # Tentukan kekuatan korelasi
    abs_corr = abs(corr)
    if abs_corr <= 0.2:
        strength = "sangat lemah"
    elif abs_corr <= 0.4:
        strength = "lemah"
    elif abs_corr <= 0.6:
        strength = "sedang"
    elif abs_corr <= 0.9:
        strength = "kuat"
    else:
        strength = "sangat kuat"

    # Tampilkan hasil analisis
    print(f"Korelasi antara '{col}' dan 'Profit' bernilai {corr:.2f}.")
    print(f"Hubungannya bersifat {direction} dengan tingkat kekuatan {strength}.")
    print("-" * 60)
```

Berdasarkan analisis korelasi Spearman terhadap variabel-variabel numerik yang berkaitan dengan profit, diperoleh beberapa temuan penting:
1. `Sales (Penjualan)` menunjukkan korelasi positif terhadap profit dengan nilai sebesar 0.23. Meskipun hubungan ini bersifat positif, kekuatannya tergolong lemah, yang berarti peningkatan penjualan hanya sedikit berpengaruh terhadap peningkatan profit.
2. `Discount (Diskon)` memiliki korelasi negatif yang cukup kuat terhadap profit dengan nilai -0.83. Ini menunjukkan bahwa semakin besar diskon yang diberikan, maka profit cenderung menurun secara signifikan. Korelasi yang kuat dan arah negatif ini mengindikasikan bahwa strategi diskon perlu dikaji ulang agar tidak menggerus keuntungan perusahaan.
3. `Profit Margin` memperlihatkan korelasi positif yang sangat kuat terhadap profit, dengan nilai 0.93. Hal ini menunjukkan bahwa peningkatan margin keuntungan sangat berkontribusi terhadap naiknya profit. Maka, meningkatkan efisiensi harga dan pengendalian biaya menjadi strategi penting untuk mengoptimalkan profitabilitas.
4. `Quantity (Jumlah Produk Terjual)` memiliki korelasi negatif yang sangat lemah terhadap profit, dengan nilai -0.05. Artinya, jumlah unit yang terjual hampir tidak berpengaruh terhadap profit.
